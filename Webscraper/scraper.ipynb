```python
import requests
from requests import exceptions
from bs4 import BeautifulSoup
# from books import Book
import pandas as pd
```

```python
def create_list(*args):
    new_list = []
    for item in args:
        new_list.append(item)
    return new_list
```

```python
all_books_details = []
# info = []
```

```python
for index in range(1, 21):
    url = f"http://books.toscrape.com/catalogue/page-{index}.html"
    try:
        response = requests.get(url=url)
    except exceptions.ConnectTimeout:
        print("Connection Timed out...")
    except exceptions.HTTPError:
        print("Error")
    else:
        soup = BeautifulSoup(response.text, "html.parser")
    products_info = soup.find("section").find("ol").find_all("li")
    for book in products_info:
        title = book.find("h3").find("a").get("title")
        price = book.find(class_="product_price").find("p").get_text().removeprefix("Â")
        link = book.find("h3").find("a").get("href")
        complete_link = "".join(["http://books.toscrape.com/catalogue/", link])
        all_books_details.append(create_list(title, price, complete_link))
        # all_books_details.append(info)

    # pprint(all_books_details)
    # str.rem
```

```python
from pprint import pprint
pprint(all_books_details)
```

```python
df = pd.DataFrame(all_books_details, columns=["Title", "Price", "Url"])
```
